```python
import pandas as pd
url="https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv"
df=pd.read_csv(url)
df
```

```python
#Para mostrar las primeras 5 rows usamoos la función dataframe.head()
print("The first 5 rows of the dataframe") 
df.head(5)
```

```python
#Para mostras las últimas 10 columnas usamos la funcion dataframe.tail()
print("The last 10 rows of the dataframe\n")
df.tail(10)
```

Para obtener un resumen estadístico de cada columna usamos el describe method
